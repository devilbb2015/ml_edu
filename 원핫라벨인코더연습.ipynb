```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data/banklist.csv')
data
```

```python
st = data.loc[:,'ST']
st
```

```python
bn = data.loc[:, 'Bank Name']
city = data.loc[:, 'City']
bn
city
```

```python
encoder = LabelEncoder()
```

```python
encoder.fit_transform(st)
```

```python
df = pd.DataFrame({'Bank_Name': bn, 'City': city})
df
```

```python
columns = df.columns
columns
```

```python
for col in columns:
    df[col] = encoder.fit_transform(df[col])
df
```
