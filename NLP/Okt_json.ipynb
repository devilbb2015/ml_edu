```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from collections import Counter
from functools import reduce
```

```python
stopWords = set(stopwords.words("english"))
lemma = WordNetLemmatizer()
```

```python
words = []

WordsToken = word_tokenize(all_data.lower())
WordsTokenStopLemma = [lemma.lemmatize(w) for w in WordsToken]
words.append(WordsTokenStopLemma)
words
```

```python
words2 = reduce(lambda x,y: x + y, words)
print(type(words2))
words2[:10]
```

```python
count = Counter(words2)
count
```

```python
word_count = dict()
```

```python
count.most_common(10)
```

```python
for tag, counts in count.most_common(10):
    if (len(str(tag)) > 1):
        word_count[tag] = counts
print(word_count)
```

```python
plt.rc('font', family='Malgun Gothic')
```

```python
plt.figure(figsize=(12,5))
plt.xlabel('키워드')
plt.ylabel('빈도수')
plt.grid(True)

plt.bar(range(len(word_count)), word_count.values())
plt.xticks(range(len(word_count)), word_count.keys(), rotation='85')
plt.show()
```
