```python
all_data = ""
```

```python
for item in json_data['items']:
    if 'description' in item.keys():
        all_data = all_data + re.sub(r'[^\w]', ' ', item['description']) + " "
print(all_data)
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from collections import Counter
from functools import reduce
```

```python
stopWords = set(stopwords.words("english"))
lemma = WordNetLemmatizer()
```

```python
words = []

WordsToken = word_tokenize(all_data.lower())
WordsTokenStopLemma = [lemma.lemmatize(w) for w in WordsToken]
words.append(WordsTokenStopLemma)
words
```

```python
words2 = reduce(lambda x,y: x + y, words)
print(type(words2))
words2[:10]
```

```python
count = Counter(words2)
count
```

```python
word_count = dict()
```

```python
count.most_common(10)
```

```python
for tag, counts in count.most_common(40):
    if (len(str(tag)) > 1):
        word_count[tag] = counts
print(word_count)
```
